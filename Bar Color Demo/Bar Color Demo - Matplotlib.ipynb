```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

cars = ['Chevy', 'BMW', 'Ford', 'Audi']
counts = [30, 10, 60, 45]

bar_labels = ['blue', 'green', 'red', 'orange']
bar_colors = ['tab:blue', 'tab:green', 'tab:red', 'tab:orange']

ax.bar(cars, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Cars Supply')
ax.set_title('Cars supply by kind and color')
ax.legend(title='Cars color')

plt.show()
```
